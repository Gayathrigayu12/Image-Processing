### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Open 'image2.jpg' from the data folder using open cv and display it in the notebook. Make sure to correct for the RGB order

```python
img=cv2.imread('image2.jpg')
```

```python
img
```

```python
plt.imshow(img)
```

### After correction

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_rgb
```

```python
plt.imshow(img_rgb)
```

### Draw an empty RED rectangle around toms face and display the image in the notebook

```python
Draw_rec=cv2.rectangle(img_rgb,pt1=(300,400),pt2=(600,700),color=(255,0,0),thickness=8)
```

```python
plt.imshow(Draw_rec)
```

### Draw a text named 'TOM' above toms face

```python
Font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
draw_text=cv2.putText(img_rgb,text='TOM',org=(500,780),fontFace=Font,fontScale=3,color=(0,0,0),thickness=10,lineType=cv2.LINE_AA)
```

```python
plt.imshow(draw_text)
```

### Do the same steps for 'Jerry'

```python
Draw_rec1=cv2.rectangle(img_rgb,pt1=(100,500),pt2=(300,700),color=(255,0,0),thickness=8)
```

```python
plt.imshow(Draw_rec1)
```

```python
Font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
draw_text1=cv2.putText(Draw_rec1,text='JERRY',org=(120,750),fontFace=Font,fontScale=2,color=(0,0,0),thickness=10,lineType=cv2.LINE_AA)
```

```python
plt.imshow(draw_text1)
```

### Convert the image to grayscale

```python
img_rgb=cv2.cvtColor(draw_text1,cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img_rgb,cmap='gray')
```
